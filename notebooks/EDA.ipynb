# Exploratory Data Analysis - Customer Churn

This notebook explores the IBM Telco Customer Churn dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_theme(style="whitegrid")
```

```python
# Load dataset
df = pd.read_csv('../data/Telco-Customer-Churn.csv')
df.head()
```

### Dataset Overview

```python
df.info()
```

### Missing Values
`TotalCharges` has empty strings that need to be converted to numeric.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print("Missing values per column:")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

### Target Class Distribution

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df, palette='Set2')
plt.title('Distribution of Customer Churn')
plt.show()
```

### Feature Distributions

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df[df['Churn'] == 'No']['tenure'], color='blue', kde=True, label='No Churn', ax=axes[0])
sns.histplot(df[df['Churn'] == 'Yes']['tenure'], color='red', kde=True, label='Churn', ax=axes[0])
axes[0].set_title('Tenure Distribution by Churn')
axes[0].legend()

sns.histplot(df[df['Churn'] == 'No']['MonthlyCharges'], color='blue', kde=True, label='No Churn', ax=axes[1])
sns.histplot(df[df['Churn'] == 'Yes']['MonthlyCharges'], color='red', kde=True, label='Churn', ax=axes[1])
axes[1].set_title('Monthly Charges Distribution by Churn')
axes[1].legend()

plt.tight_layout()
plt.show()
```

### Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
# Convert Churn to numeric for correlation
df_corr = df.copy()
df_corr['Churn'] = df_corr['Churn'].map({'Yes': 1, 'No': 0})

corr = df_corr[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
